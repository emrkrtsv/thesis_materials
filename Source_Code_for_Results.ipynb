ANALYSIS OF THE EXCEL FILE KEEPING SCENARIOS AND FORMATS STABLE

```python
#downloading the file
file_path = 'InclusiveG-EN_2.0.xlsx'
df = pd.read_excel(file_path, sheet_name='Aggregated')
```

```python
# 2. Definition of words with their code in the excel file
model_map = {
    'M1': 'ChatGPT 5.2',
    'M2': 'Gemini Pro 3',
    'M3': 'Claude Sonnet 4.5'
}
model_order = [model_map['M1'], model_map['M2'], model_map['M3']]

scenario_map = {
    'S1': 'Lexical Gender',
    'S2': 'Referential Gender',
    'S3': 'Occupational Gender',
    'S4': 'Adjectival /Descriptive Gender',
    'S5': 'Implicit Gender',
    'S6': 'Pragmatic Gender'
}

form_map = {
    'F1': 'Rewriting',
    'F2': 'Structure-preserving Paraphrasing'
}

technique_map = {
    'T1': 'Zero-shot',
    'T2': 'Few-shot',
    'T3': 'Rule-based',
    'T4': 'Chain-of-Thought',
    'T5': 'Constraint-based Adversarial'
}
```

```python
# Creating the columns
# Dividing the 'M-S-F-T' structure.
df[['Model_ID', 'Scenario_ID', 'Form_ID', 'Technique_ID']] = df['Output_ID'].str.split('-', expand=True)
df['Mean_Final_Score_0_15'] = pd.to_numeric(df['Mean_Final_Score_0_15'], errors='coerce')

# Converting ID's to names
df['Model'] = df['Model_ID'].map(model_map)
df['Scenario'] = df['Scenario_ID'].map(scenario_map)
df['Form'] = df['Form_ID'].map(form_map)
df['Technique'] = df['Technique_ID'].map(technique_map)

# All combinations to be analyzed
combinations = [
    ('S1', 'F1'), ('S1', 'F2'), ('S2', 'F1'), ('S2', 'F2'),
    ('S3', 'F1'), ('S3', 'F2'), ('S4', 'F1'), ('S4', 'F2'),
    ('S5', 'F1'), ('S5', 'F2'), ('S6', 'F1'), ('S6', 'F2')
]
```

```python
#improvement calculation - using T1 as a reference
def calculate_improvement(group):
    t1_row = group.loc[group['Technique_ID'] == 'T1', 'Mean_Final_Score_0_15']
    if not t1_row.empty:
        t1_score = t1_row.values[0]
        if t1_score != 0:
            group['The Percentage of Improvement'] = ((group['Mean_Final_Score_0_15'] - t1_score) / t1_score) * 100
        else:
            group['The Percentage of Improvement'] = 0
    else:
        group['The Percentage of Improvement'] = 0
    return group

import matplotlib.pyplot as plt
import seaborn as sns

#Analysis and Visualization Loop
for sce_id, frm_id in combinations:
    current_sce = scenario_map.get(sce_id)
    current_frm = form_map.get(frm_id)

    # Filtering the data
    target_df = df[(df['Scenario_ID'] == sce_id) & (df['Form_ID'] == frm_id)].copy()
    if target_df.empty:
        continue

    # Calculate improvements using your existing function
    final_table = target_df.groupby('Model', group_keys=False).apply(calculate_improvement)

    # Creating a barplot and a table side to side
    # 1 row, 2 columns. gridspec_kw adjusts the relative width of plot vs table
    fig, (ax_plot, ax_table) = plt.subplots(1, 2, figsize=(20, 8), gridspec_kw={'width_ratios': [1.6, 1]})

    sns.set_style("whitegrid")
    custom_palette = {model_map['M1']: 'seagreen', model_map['M2']: 'royalblue', model_map['M3']: 'crimson'}
    tech_order = [technique_map[t] for t in ['T1', 'T2', 'T3', 'T4', 'T5'] if t in technique_map]

    # LEFT PANEL: BARPLOT
    sns.barplot(data=final_table, x='Technique', y='Mean_Final_Score_0_15',
                hue='Model', palette=custom_palette, hue_order=model_order, order=tech_order, ax=ax_plot)

    ax_plot.set_title(f'Performance Comparison\n Scenario: {current_sce} - Form: {current_frm}', fontsize=14, fontweight='bold')
    ax_plot.set_ylim(0, 18)
    ax_plot.tick_params(axis='x', rotation=15)
    ax_plot.set_ylabel('Mean Final Score (0-15)')
    ax_plot.legend(title='Models', loc='upper left')

    # Annotate bar heights
    for p in ax_plot.patches:
        if p.get_height() > 0:
            ax_plot.annotate(format(p.get_height(), '.2f'),
                             (p.get_x() + p.get_width() / 2., p.get_height()),
                             ha='center', va='center', xytext=(0, 9),
                             textcoords='offset points', fontsize=9, fontweight='bold')

    # RIGHT PANEL: TABLE PREPARATION
    detailed_imp = final_table.pivot(index='Technique', columns='Model', values='The Percentage of Improvement')
    detailed_imp = detailed_imp.reindex(tech_order)
    existing_models = [m for m in model_order if m in detailed_imp.columns]
    detailed_imp = detailed_imp[existing_models]

    display_data = detailed_imp.reset_index()
    #adding a two-line code to fix the problem with "Constraint-based Adversarial"
    display_data['Technique'] = display_data['Technique'].apply(
    lambda x: x.replace("Constraint-based Adversarial", "Constraint-based\nAdversarial")
)
    for col in existing_models:
        display_data[col] = display_data[col].apply(lambda x: f"{x:.2f}%" if pd.notnull(x) else "0.00%")

    # Drawing the table on the second axis
    ax_table.axis('off')
    the_table = ax_table.table(cellText=display_data.values,
                               colLabels=display_data.columns,
                               cellLoc='center',
                               loc='center')

    # Table styling
    the_table.auto_set_font_size(False)
    the_table.set_fontsize(10)
    the_table.scale(1, 3.0) # Increases cell height for better readability

    # Coloring logic
    for (row, col), cell in the_table.get_celld().items():
        if row == 0:
            cell.set_facecolor("#2c3e50")
            cell.get_text().set_color('white')
            cell.get_text().set_weight('bold')
        elif row > 0 and col > 0:
            cell_text = str(cell.get_text().get_text())
            if "-" in cell_text:
                cell.get_text().set_color('red') # Red for negative improvement
            else:
                cell.get_text().set_color('black')

    ax_table.set_title(f'Improvement % (Relative to Zero-shot Prompting)', fontsize=12, fontweight='bold', pad=20)

    plt.tight_layout()
    plt.show()
```
